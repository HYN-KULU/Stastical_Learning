```python
import pandas
import torch

data = pandas.read_csv('../C_Data.csv')
length = len(data['y1'])
s1 = torch.tensor(data['p1']).reshape((length, 1))
s2 = torch.tensor(data['p2']).reshape((length, 1))
s3 = torch.tensor(data['p3']).reshape((length, 1))
s4 = torch.tensor(data['p4']).reshape((length, 1))
s5 = torch.tensor(data['p5']).reshape((length, 1))
s6 = torch.tensor(data['p6']).reshape((length, 1))
s7 = torch.tensor(data['p7']).reshape((length, 1))

a1 = torch.tensor(data['t1']).reshape((length, 1))
a2 = torch.tensor(data['t2']).reshape((length, 1))
a3 = torch.tensor(data['t3']).reshape((length, 1))
a4 = torch.tensor(data['t4']).reshape((length, 1))
a5 = torch.tensor(data['t5']).reshape((length, 1))
a6 = torch.tensor(data['t6']).reshape((length, 1))
a7 = torch.tensor(data['t7']).reshape((length, 1))

input = torch.cat((s1, s2, s3, s4, s5, s6, s7), 1).to(torch.float32).to('cuda')
output = torch.cat((a1, a2, a3, a4, a5, a6, a7),
                   1).to(torch.float32).to('cuda')
```

```python
import torch

# nn2 = 10
net1 = torch.nn.Sequential(
    torch.nn.Linear(7,100),
    torch.nn.ReLU(),
    torch.nn.Linear(100,100),
    torch.nn.ReLU(),
    torch.nn.Linear(100,7),
    ).to('cuda')
x = input
y = output
```

```python
optimizer = torch.optim.SGD(net1.parameters(), lr=0.001)
loss_function = torch.nn.MSELoss()
for epoch in range(1000000):
   y_pred = net1(x)
   loss = loss_function(y_pred, y)
   print('epoch: ', epoch,' loss: ', loss.item())
   optimizer.zero_grad()
   loss.backward()
   optimizer.step()
```
