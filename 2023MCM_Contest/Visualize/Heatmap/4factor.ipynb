```python
import numpy
q=numpy.loadtxt(open("../../correlation/PCA_corrcoef.csv", "rb"), delimiter=",", skiprows=0)
cmatrix = q.tolist()
```

```python
from pyecharts import options as opts
from pyecharts.charts import HeatMap

key = ['Vowel', 'Overlay', 'Frequency', 'Entropy']
value = [[i, j,'%.2f' % cmatrix[i][j]] for i in range(4) for j in range(4)]
c = (
    HeatMap()
    .add_xaxis(key)
    .add_yaxis(
        "",
        key,
        value,
        label_opts=opts.LabelOpts(is_show=True, position="inside"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="4 Factor Correlation"),
        visualmap_opts=opts.VisualMapOpts(min_=-1, max_=1, is_calculable=True, orient="horizontal", pos_left="center"),
        
    )
    .render("result.html")
)
```

```python
from pyecharts import options as opts
from pyecharts.charts import HeatMap

key = ['Mean', 'Variance', 'X1', 'X2', 'X3', 'X4', 'X5', 'Y1', 'Y2', 'Y3']
length = len(cmatrix)
value = [[i, j,'%.2f' % cmatrix[i][j]] for i in range(length) for j in range(length)]
c = (
    HeatMap()
    .add_xaxis(key)
    .add_yaxis(
        "",
        key,
        value,
        label_opts=opts.LabelOpts(is_show=True, position="inside"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="4 Factor Correlation"),
        visualmap_opts=opts.VisualMapOpts(min_=-1, max_=1, is_calculable=True, orient="horizontal", pos_left="center"),
        
    )
    .render("result.html")
)
```
