```python
from jmutils.EWAnalysis.EWAnalysis import EWeight
import numpy as np
import pandas as pd

data = pd.read_csv('./C_Data.csv')

# for _, row in data.iterrows():
#     print(row[1])

# 1 try,2 tries,3 tries,4 tries,5 tries,6 tries,7 or more tries (X)
s1 = np.array(data['1 try']).reshape((356,1))
s2 = np.array(data['2 tries']).reshape((356,1))
s3 = np.array(data['3 tries']).reshape((356,1))
s4 = np.array(data['4 tries']).reshape((356,1))
s5 = np.array(data['5 tries']).reshape((356,1))
s6 = np.array(data['6 tries']).reshape((356,1))
s7 = np.array(data['7 or more tries (X)']).reshape((356,1))

score = np.concatenate([s1,s2,s3,s4,s5,s6,s7], axis=1) + 0.01
```

```python
ew = EWeight(score)
ew.run()
ew.printresult()
```
