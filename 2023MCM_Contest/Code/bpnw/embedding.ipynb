```python
import pandas
import torch

data = pandas.read_csv('./C_Data_Train.csv')

length = len(data['y1'])
s1 = torch.tensor(data['p1']).reshape((length,1))
s2 = torch.tensor(data['p2']).reshape((length,1))
s3 = torch.tensor(data['p3']).reshape((length,1))
s4 = torch.tensor(data['p4']).reshape((length,1))
s5 = torch.tensor(data['p5']).reshape((length,1))
# q1 = torch.tensor(data['y1']).reshape((length,1))
# q2 = torch.tensor(data['y2']).reshape((length,1))
# q3 = torch.tensor(data['y3']).reshape((length,1))

a1 = torch.tensor(data['mu']).reshape((length,1))
a2 = torch.tensor(data['sigma']).reshape((length,1))

input = torch.cat((s1,s2,s3,s4,s5),1).to(torch.float32).to('cuda')
output = torch.cat((a1,a2),1).to(torch.float32).to('cuda')

```

```python
data = pandas.read_csv('../C_Data.csv')

def word_to_nlist(word):
    sk = word
    nl = []
    for i in sk:
        nl.append(str(ord(i)-97))
    return nl

for word in data['Word']:
    with open('code.csv', 'a') as f:
        f.write(",".join(word_to_nlist(word)) + '\n')
```

```python
import torch

net1 = torch.nn.Sequential(
    torch.nn.Linear(5,5),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(5,2)
    ).to('cuda')
x = input
y = output
```

```python
optimizer = torch.optim.SGD(net1.parameters(), lr=0.1)
loss_function = torch.nn.MSELoss()
for epoch in range(1000):
   y_pred = net1(x)
   loss = loss_function(y_pred, y)
   print('epoch: ', epoch,' loss: ', loss.item())
   optimizer.zero_grad()
   loss.backward()
   optimizer.step()
```
